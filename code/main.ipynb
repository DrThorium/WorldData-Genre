```python
# imports
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# route of the data base

dataBaseRoute = "../source/WDIData.csv"
countriesRoute = "../source/WDICountry.csv"
```

```python
df = pd.read_csv(dataBaseRoute)
df.head(10)
```

```python
country = pd.read_csv(countriesRoute)
country.head(10)
```

```python
# Select only countries from Latin America and Income Group = Upper middle income

# Preselect only the columns that we need

country = country[['Country Code', 'Region', 'Income Group']]

latinAmerica = country[country['Region'] == 'Latin America & Caribbean']

latinAmerica = latinAmerica[latinAmerica['Income Group'] == 'Upper middle income']

# Extract the country codes to a list

latinAmericaCodes = latinAmerica['Country Code'].tolist()

# Delete islands from the list

latinAmericaCodes.remove('GRD')
latinAmericaCodes.remove('JAM')
latinAmericaCodes.remove('LCA')
latinAmericaCodes.remove('VCT')
latinAmericaCodes.remove('SUR')
latinAmericaCodes.remove('DMA')



```

```python
# Filter the data base with the country codes

df = df[df['Country Code'].isin(latinAmericaCodes)]
df.head(10)
```

```python
# Check data availability for each indicator, for each country, for each year using a pipeline

# Check the list of indicators

indicators = df['Indicator Name'].unique().tolist()

# Check the list of countries

countries = df['Country Name'].unique().tolist()




# The list of years is all columns next to Indicator Code and before the last column

years = df.columns.tolist()
years = years[4:-1]

# delete the last column from df

df = df.iloc[:, :-3]

# Select only the first 4 columns and the columns with the years 2000 to 2020. 

df_4 = df.iloc[:, :4]
df_last = df.iloc[:, -20:]

df = pd.concat([df_4, df_last], axis=1)

```

```python
# Melt the data base to have the Country Name, Country Code, Indicator Name, Indicator Code, Year and the Indicators as columns

df = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')
```

```python
# Pivot the data base to have all the indicators as columns

df = df.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns='Indicator Name', values='Value')

# Reset the index

df = df.reset_index()

# Check the data base

df.head(10)
```

```python
# Find the most problematic indicators

# Check the number of null values for each indicator

df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Delete the indicators with more than 20% null values using trehshold = 0.2

df = df.dropna(thresh=len(df)*0.8, axis=1)

# Check the number of null values for each indicator

df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Because the year is not a useful variable to predict the GDP, we will delete it
df_original = df.copy()




df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2020')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2019')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2018')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2017')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2016')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2015')].index)
df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2014')].index)
# df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2013')].index)
# df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2012')].index)
# df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2011')].index)
# df = df.drop(df[(df['Country Name'] == 'Peru') & (df['Year'] == '2010')].index)


df = df.drop(['Year'], axis=1)
```

```python
# Check the total of indicators

df.shape
```

```python
# Delete all indicators containing the word "GDP" except the GDP Current US$

# Rename the column GDP Current US$ to "Y"

df = df.rename(columns={'GDP (current US$)': 'Y'})

df = df.drop(df.filter(regex='GDP').columns, axis=1)

# Check the total of indicators

df.shape
```

```python

df_income = country[['Country Code', 'Income Group']]
df_income = df_income.dropna()

# Merge the data base with the income group keep only Non-null values
df_train_Income = df.merge(df_income, how='inner', on='Country Code')
df_train_Income = df_train_Income.drop(['Country Name', 'Country Code'], axis=1)
df_train_Income = df_train_Income.drop(['Y'], axis=1)

df_train_Income.head(10)

# Separete X, y, X_test and y_test

X_income = df_train_Income.drop(['Income Group'], axis=1)
y_income = df_train_Income['Income Group']

# Split the data base into train and test

X_train_income, X_test_income, y_train_income, y_test_income = train_test_split(X_income, y_income, test_size=0.2, random_state=42)

```

```python
# make a df for training (all columns except Y) and a df for testing (only Y)


df_train = df.drop(['Y'], axis=1)

# Drop Country Name and Country Code

df_train = df_train.drop(['Country Name', 'Country Code'], axis=1)

df_test = df[['Y']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

# Supongamos que df_train es tu DataFrame con valores NaN
# Crea un imputador SimpleImputer con estrategia 'mean' para manejar NaN inicialmente
imputer = SimpleImputer(strategy='mean', keep_empty_features=True)
KNNImputer = KNNImputer(n_neighbors=2, weights="uniform", keep_empty_features=True)

KNNImputer.fit(df_train)

imputer.fit(df_train)

# Llena los valores NaN en df_train utilizando el imputador
df_train_imputed = imputer.fit_transform(df_train)

# Convierte el resultado de vuelta a un DataFrame
df_train_imputed = pd.DataFrame(df_train_imputed, columns=df_train.columns)

df_train_imputed.head(10)
```

```python

# combine the training and testing data frames

df = pd.concat([df_train_imputed, df_test], axis=1)

# get X, y


X = df.drop(['Y'], axis=1)
indicators = df_train.columns.tolist()
indicators
y = df[['Y']]
y = y.values.ravel()

# Adjust scale of X

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
# Use PCA to reduce the number of features

from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)

X = pca.fit_transform(X)
```

```python
# Matriz de cargas usando Indicator Name para identificar las variables mejor


loadings = pd.DataFrame(pca.components_.T)

loadings['Indicator Name'] = df_train.columns
```

```python
loadings
```

```python
# Get top 10 indicators with highest loadings in PC1, using absolute value



loadings[0] = abs(loadings[0])
indicators_names = loadings.sort_values(by=0, ascending=False).head(20)['Indicator Name'].tolist()
```

```python
def top_variables_for_components(loadings):
    # Crear una copia de loadings para no modificar la matriz original
    loadings_copy = loadings.copy()
    
    # Tomar el valor absoluto de los loadings
    loadings_copy.iloc[:, :-1] = abs(loadings_copy.iloc[:, :-1])
    
    # Inicializar un diccionario para almacenar los resultados
    top_variables_dict = {}
    
    # Obtener el número de componentes principales (excluyendo la columna de nombres)
    num_components = loadings_copy.shape[1] - 1
    
    for i in range(num_components):
        # Seleccionar los 20 mejores indicadores para el componente i
        top_indicators = loadings_copy.sort_values(by=i, ascending=False).head(20)['Indicator Name'].tolist()
        
        # Agregar la lista de los 20 mejores indicadores al diccionario
        top_variables_dict[f"Componente {i}"] = top_indicators
    
    return top_variables_dict


results = top_variables_for_components(loadings)

print(results)






```

```python
# Transform results into a data frame

results = pd.DataFrame(results)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2718281828, test_size=0.2)

lr = LinearRegression()
nn = MLPRegressor(hidden_layer_sizes=(100, 50, 100), random_state=2718281828, max_iter=10000, learning_rate_init=0.2)

lr.fit(X_train, y_train)
nn.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred_nn = nn.predict(X_test)


from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
mse_nn = mean_squared_error(y_test, y_pred_nn)

print(f"El error cuadrático medio es: {mse}")
print(f"El error cuadrático medio de la red neuronal es: {mse_nn}")


from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
r2_nn = r2_score(y_test, y_pred_nn)

print(f"El coeficiente de determinación es: {r2}")
print(f"El coeficiente de determinación de la red neuronal es: {r2_nn}")
```

```python
# Create a pipeline to predict the GDP of a country with the next steps:

# 1. With a given country, extract the indicators from the data base.
# 2. Execute the imputer to fill the NaN values.
# 3. Execute the scaler to adjust the scale of the indicators.
# 4. Execute PCA to reduce the number of indicators.
# 5. Execute the model to predict the GDP.
# 6. Return the predicted GDP and scores.

# Create a function to execute the pipeline
df = pd.read_csv(dataBaseRoute)

def predict_gdp(country_code, df_original_, indicators, imputer, scaler, pca, model):
    indicators_ = indicators.copy()
    indicators_.append('GDP (current US$)')
    # Select only the country of interest:
    df_country = df_original_[df_original_['Country Code'] == country_code]
    df_country = df_country[df_country['Indicator Name'].isin(indicators_)]
    df_country = df_country.rename(columns={'GDP (current US$)': 'Y'})
    years = df_country.columns.tolist()
    years = years[4:-1]
    # delete the last column from df
    df_country = df_country.iloc[:, :-3]
    # Select only the first 4 columns and the columns with the years 2000 to 2020. 
    df_4 = df_country.iloc[:, :4]
    df_last = df_country.iloc[:, -20:]
    df_country = pd.concat([df_4, df_last], axis=1)
    # Melt the data base to have the Country Name, Country Code, Indicator Name, Indicator Code, Year and the Indicators as columns
    df_country = df_country.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')
    # Pivot the data base to have all the indicators as columns
    df_country = df_country.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns='Indicator Name', values='Value', dropna=False)
    # Reset the index
    df_country = df_country.reset_index()
    # Delete all indicators containing the word "GDP" except the GDP Current US$
    df_country = df_country.rename(columns={'GDP (current US$)': 'Y'})
    df_country_Y = df_country[['Y']]
    # Drop "Y" Column
    df_country = df_country.drop(['Y'], axis=1) 
    # Select only indicators used in the model and Y column
    # Delete the year column
    df_country = df_country.drop(['Year'], axis=1)
    # Delete Country Name and Country Code
    df_country = df_country.drop(['Country Name', 'Country Code'], axis=1)
    # Use the imputer to fill the NaN values
    df_country = imputer.transform(df_country)
    # Use the scaler to adjust the scale of the indicators
    df_country = scaler.transform(df_country)
    # Use PCA to reduce the number of indicators
    df_country = pca.transform(df_country)
    # Use the model to predict the GDP
    y_pred = model.predict(df_country)
    # Scores of the model y_pred vs df_country_Y
    r2 = r2_score(df_country_Y, y_pred)
    # Create a data frame with the results comparing the real GDP and the predicted GDP
    results = pd.DataFrame({'Real GDP': df_country_Y['Y'], 'Predicted GDP': y_pred})
    # Calculate the error
    results['Error'] = results['Real GDP'] - results['Predicted GDP']
    # Add the year column
    # Return the results
    return results, r2

# Test the function with Peru

y_pred = predict_gdp('CHL', df, indicators, imputer, scaler, pca, lr)
print(y_pred)
```
